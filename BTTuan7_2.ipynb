```python
#W7A16
a = list(map(int, input().split()))
freq = {}
for x in a:
    if x in freq:
        freq[x] += 1
    else:
        freq[x] = 1
        
remove = 0
for x in freq:
    cnt = freq[x]

    if cnt == x:
        continue
    elif cnt > x:
        remove += cnt - x
    else:
        remove += cnt

if remove == 0:
    print("YES")
else:
    print("NO", remove)
```

```python
#W7A19
a = list(map(int, input().split()))
b = list(map(int, input().split()))
i = j = 0
res = []

ascending = True
if len(a) >= 2 and a[0] > a[-1]:
    ascending = False

while i < len(a) and j < len(b):
    if ascending:
        if a[i] <= b[j]:
            res.append(a[i])
            i += 1
        else:
            res.append(b[j])
            j += 1
    else:  # giảm dần
        if a[i] >= b[j]:
            res.append(a[i])
            i += 1
        else:
            res.append(b[j])
            j += 1

while i < len(a):
    res.append(a[i])
    i += 1
while j < len(b):
    res.append(b[j])
    j += 1

print(*res)
```

```python
#W7A22
n = int(input())
freq = {}
for _ in range(n):
    full_name = input().strip()
    first_name = full_name.split()[-1] 
    freq[first_name] = freq.get(first_name, 0) + 1

best = ""
max_freq = 0
for name in freq:
    if freq[name] > max_freq or (freq[name] == max_freq and name < best):
        max_freq = freq[name]
        best = name
print(best)
```

```python
#W7A25
s = input().strip()
k = int(input())

remove = len(s) - k
stack = []
for d in s:
    while stack and remove > 0 and stack[-1] < d:
        stack.pop()
        remove -= 1
    stack.append(d)
result = "".join(stack[:k])
print(result)
```

```python
#W7A27
s = input()
stack = []
match = {
    ')': '(',
    '}': '{',
    ']': '['
}

for ch in s:
    if ch in "({[":
        stack.append(ch)
    else:
        if not stack or stack[-1] != match[ch]:
            print("NO")
            break
        stack.pop()
else:
    if not stack:
        print("YES")
    else:
        print("NO")
```

```python
#W7A28
N = int(input())
def is_prime(x):
    if x < 2:
        return False
    for i in range(2, int(x**0.5)+1):
        if x % i == 0:
            return False
    return True

max_len = 0
best_start = 0

cur_len = 0
cur_start = 0

for i in range(2, N+1):
    if not is_prime(i): 
        if cur_len == 0:
            cur_start = i
        cur_len += 1
    else:
        if cur_len > max_len:
            max_len = cur_len
            best_start = cur_start
        cur_len = 0

if cur_len > max_len:
    max_len = cur_len
    best_start = cur_start

for i in range(best_start, best_start + max_len):
    print(i, end=" ")
```

```python
#W7A29
a = list(map(int, input().split()))

best = start = 0
maxlen = cur = 1

for i in range(1, len(a)):
    cur = cur+1 if a[i] > a[i-1] else 1
    start = start if a[i] > a[i-1] else i
    if cur > maxlen:
        maxlen, best = cur, start

print(*a[best:best+maxlen])
```

```python
#W7A30
a = list(map(int, input().split()))
n = len(a)
dp = [1]*n
par = [-1]*n

for i in range(n):
    for j in range(i):
        if a[j] < a[i] and dp[j]+1 > dp[i]:
            dp[i] = dp[j]+1
            par[i] = j

idx = max(range(n), key=lambda i: dp[i])

res = []
while idx != -1:
    res.append(a[idx])
    idx = par[idx]

print(*res[::-1])
```

```python
#W7A31
s = list(input().strip())
n = len(s)
for i in range(n):
    pos = -1
    mn = s[i]

    for j in range(n-1, i, -1):
        if s[j] < mn:
            mn = s[j]
            pos = j

    if pos != -1 and not (i == 0 and s[pos] == '0'):
        s[i], s[pos] = s[pos], s[i]
        break
print(''.join(s))
```
