```python
#Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng
#giữ nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả.
def rmv(A):
    L = []
    for i in A:
        if i not in L:
            L.append(i)
    return L
A = list(map(int, input().split()))
print(rmv(A))


```

```python
#Nhập một tuple số nguyên (một dòng) và số nguyên k (dòng kế). In ra tuple
#sau khi xoay trái k vị trí. (Giữ nguyên thứ tự tương đối sau xoay; không sắp
#xếp.)
A=tuple(map(int,input().split()))
k=int(input())
B=list(A)
n=len(B)
C=[]
for i in range (k,n):
    C.append(B[i])
for i in range (0,k):
    C.append(B[i])
print(tuple(C))
```

```python
#Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo
#dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key
dn=input().split()
dic={}
for i in dn :
    key,value=i.split(':')
    a=[]
    if key not in dic :
        a.append(value)
        dic[key]=a
    else:
        dic[key].append(value)
print(dic)
```

```python
#Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một
#dictionary mới trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ
#được cộng lại, nếu chỉ có trong một dict thì giữ nguyên. In ra dictionary mới
#theo thứ tự từ điển tăng dần của key. 
nhap1=list(input().split())
nhap2=list(input().split())
dict={}
for phantu in nhap1:
    key,val=phantu.split(':')
    dict[key]=val
for phantu in nhap2:
    key,val=phantu.split(':')
    if key in dict :
        dict[key]=int(dict[key])+int(val)
    else:
        dict[key]=val
print(dict)
```

```python
#Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên
#dương k. In ra list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là
#một tuple, hai số trong tuple có thể giống nhau, thứ tự xuất hiện trong danh
#sách của số thứ nhất nhỏ hơn thứ tự của số thứ hai. Các tuple trong list được
#sắp xếp theo thứ tự tăng dần của phần tử đầu tiên trong tuple.
nhap=list(map(int,input().split()))
k=int(input())
A=[]
for i in range (len(nhap)):
 for j in range (i+1,len(nhap)):
    if nhap[i]+nhap[j]==k : 
        A.append((nhap[i],nhap[j]))
print(A)
```

```python
#[Tách số chẵn và lẻ từ tuple]
#Nhập vào một tuple chứa các số nguyên. Hãy tạo ra hai tuple mới: một chứa
#tất cả các số chẵn và một chứa tất cả các số lẻ từ tuple ban đầu, giữ nguyên
#thứ tự tương đối. In ra hai tuple này trên hai dòng riêng biệt.
A=tuple(map(int,input().split()))
B=[]
C=[]
for i in A:
    if i%2==0 :
        B.append(i)
    else:
        C.append(i)
print(B)
print(C)
```

```python
#[Tìm phần tử xuất hiện nhiều nhất]
#Nhập vào một list các số nguyên. Hãy tìm số xuất hiện nhiều lần nhất trong
#list. Nếu có nhiều số cùng có số lần xuất hiện nhiều nhất, hãy in ra số nhỏ
#nhất trong số đó
A=list(map(int,input().split()))
dic={}
#đếm số lần xuất hiện 
for i in A:
    if i in dic:
       dic[i]+=1
    else:
        dic[i]=1
#tìm số có tuần suất lớn nhất nhưng nhỏ hơn 
sln=max(dic.values())
N=[]
for k,v in dic.items() :
    if v==sln :
        N.append(k)
print(min(N))
```

```python
#[Đảo ngược key và value của dictionary]
#Nhập vào một dictionary có value là duy nhất (không trùng nhau). Hãy tạo một
#dictionary mới bằng cách đảo ngược key và value của dictionary ban đầu. In
#ra dictionary mới. 
a=list(input().split())
dic={}
#tạo dic ban đầu
for i in a : 
    key,val=i.split(":")
    dic[key]=val
#đảo ngược key và val
reverse_dic={}
for k,v in dic.items():
    reverse_dic[v]=k
print(reverse_dic)
```

```python
def most_frequent(arr):
    d={}
    for i in arr:
        if i not in d:
            d[i]=1
        else:
            d[i]+=1
    val=d.values()
    max1=max(val)
    for key in d.keys():
        if d[key]==max1:
            return key
numbers = list(map(int, input().strip().split()))
print(most_frequent(numbers))
```

```python
def classifyNumber(n):
  dem=0
  for i in range (1,n+1):
      if n%i==0 :
        dem=dem+i
  if dem < 2*n :
     return "-1"
  elif dem>2*n:
     return "1"
  else:
    return "0"
n = int(input())
print(classifyNumber(n))
```
