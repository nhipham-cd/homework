```python
#W7A1 Viết hàm tìm kiếm nhị phân trên một danh sách đã được sắp xếp tăng dần. 
#Trả về chỉ số của phần tử hoặc -1 nếu không tìm thấy. mẫu hàm: def binary_search(arr, left, right, target): 
def binary_search (arr, left, right, target):
    if left > right:
        return -1
    mid = (left + right)//2
    if arr[mid] == target:
        return mid
    elif arr[mid] < target:
        return binary_search (arr, mid+1, right, target)
    else:
        return binary_search (arr, left, mid-1, target)

arr = list(map(int,input().split()))
target = int(input())

print (binary_search(arr, 0, len(arr)-1, target))
```

```python
#W7A3
def bubble_sort(arr):
    n = len(arr)
    for i in range (n-1):
        for j in range (n-1-i):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

def selec(arr):
    n =len(arr)
    for i in range (n-1):
        min_index = i
        for j in range (i+1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return arr

arr = list(map(int, input().split()))
print("Bubble Sort:", *bubble_sort(arr.copy()))
print("Selection Sort:", *selec(arr.copy()))
```

```python
#W7A4 tim phan tu xuat hien nhieu nhat, = nhau thi in ra phan tu truoc
def most_frequent(arr):
    max_dem = 0
    so = arr[0]
    for i in range (len(arr)):
        dem = 0
        for j in range (len(arr)):
            if arr[j]==arr[i]:
                dem += 1
            if max_dem < dem:
                max_dem = dem
                so = arr[i]
    return so, max_dem

arr = list(map(int, input().split()))
so, max_dem = most_frequent(arr)
print (f"{so} xuat hien nhieu nhat, som nhat, {max_dem} lan")
```

```python
#W7A5 
def count_pairs(arr, X):
    dem = 0
    n = len(arr)

    for i in range(n):
        for j in range(i + 1, n):
            if arr[i] + arr[j] == X:
                dem += 1
    return dem

arr = list(map(int, input().split()))
X = int(input())
print(count_pairs(arr, X))
```

```python
#W7A7 day tang nghiem ngat nhat
def lis_length(nums):
    n = len(nums)
    dp = [1] * n

    for i in range(n):
        for j in range(i):
            if nums[j] < nums[i]:
                dp[i] = max(dp[i], dp[j] + 1)
    return max(dp)

nums = list(map(int, input().split()))
print(lis_length(nums))
```

```python
#7A8
import heapq

def min_interval(intervals, queries):
    intervals.sort()
    qs = sorted((q, i) for i, q in enumerate(queries))
    res = [-1] * len(queries)

    heap = []
    i = 0

    for q, idx in qs:
        while i < len(intervals) and intervals[i][0] <= q:
            l, r = intervals[i]
            heapq.heappush(heap, (r - l + 1, r))
            i += 1

        while heap and heap[0][1] < q:
            heapq.heappop(heap)

        if heap:
            res[idx] = heap[0][0]

    return res

n = int(input())
intervals = []
for _ in range(n):
    l, r = map(int, input().split())
    intervals.append([l, r])

queries = list(map(int, input().split()))
print(min_interval(intervals, queries))
```

```python
#W7A9
def dem_sonho(nums):
    res = []
    sorted_list = []

    for x in reversed(nums):
        left, right = 0, len (sorted_list)
        while left < right:
            mid = (left + right) // 2
            if sorted_list[mid] < x:
                left = mid + 1
            else:
                right = mid
        res.append(left)
        sorted_list.insert(left, x)
    return res[::-1]

nums = list(map(int, input().split()))
print(dem_sonho(nums))
```

```python
#W7A10
def max_subarray_sum_mod(nums, m):
    mang_prefix = []
    tong = 0
    ket_qua = 0

    for x in nums:
        tong = (tong + x) % m
        ket_qua = max(ket_qua, tong)

        trai, phai = 0, len(mang_prefix)
        while trai < phai:
            giua = (trai + phai) // 2
            if mang_prefix[giua] <= tong:
                trai = giua + 1
            else:
                phai = giua

        if trai < len(mang_prefix):
            ket_qua = max(ket_qua, (tong - mang_prefix[trai] + m) % m)

        mang_prefix.insert(trai, tong)
    return ket_qua

nums = list(map(int, input().split()))
m = int(input())
print(max_subarray_sum_mod(nums, m))
```
