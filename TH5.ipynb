```python
#W5A1 
#Viết hàm truyền vào hai số nguyên a, b. Trả về số lớn hơn trong hai số. def Max_Of_Two(a, b) -> int: 
def max_of_two (a: int, b: int): #kieu du lieu cua ham
    return max(a, b)

try:
    a, b = map(int, input().split())
    print (max_of_two(a, b))
except ValueError:
    print ('nhap so nguyen vao')
    
```

```python
#W5A3 Viết hàm kiểm tra một số có phải là số nguyên tố hay không? 
import math
def prime_num(n):
    if n <2: 
        return False
    for i in range (2, int(math.sqrt(n))+1):
        if n%i==0:
            return False
    return True

n = int(input())
if (prime_num(n)):
    print ("True")
else:
    print ("False")
```
