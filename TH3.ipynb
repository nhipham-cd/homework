```python
#W3A3 Kiểm tra xem một số có phải là lũy thừa của 2 hay không 
n = int(input())

if n > 0 and (n & (n - 1)) == 0:
    print("YES")
else:
    print("NO")
```

```python
#W3A6 [Even and Odd]  
#Nhập vào một số nguyên dương x, bạn hãy kiểm tra xem x là số chẵn hay lẻ. 
#Nếu x là số chẵn, in ra "Even". Nếu x là số lẻ, in ra "Odd".  
x = int(input())
if x % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
#W3A9 [Triangle]  
# Nhập vào 3 số nguyên dương a, b, c. Kiểm tra xem a, b, c có cấu thành độ dài cạnh
# của 1 tam giác được không. Nếu a, b, c cấu tạo thành được một tam giác, 
# in ra "Yes". Ngược lại in ra "No". 
a,b,c = map(int,input().split())
if a + b > c and a + c > b and b + c > a:
    print("Yes")
else:
    print("No")
```

```python
#W3A10 [Largest number] 
#Nhập vào 4 số nguyên. In ra màn hình số nguyên lớn nhất trong 4 số.
a,b,c,d = map(int,input().split())
# print(max(a, b, c, d))
max_num = a
if b > max_num:
    max_num = b
if c > max_num:
    max_num = c
if d > max_num:
    max_num = d
print(max_num)
```

```python
#W3A11 [Phân loại tam giác] 
a, b, c = map(int, input().split())
if a + b <= c or a + c <= b or b + c <= a:
    print("Không phải tam giác")
else:
    if a == b == c:
        print("Tam giác đều")
    elif a == b or a == c or b == c:
        print("Tam giác cân")
    else:
        print("Tam giác thường")
```
