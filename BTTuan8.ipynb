```python
#W8A1 Cylinder
class Cylinder:
    pi = 3.14
    def __init__(self, radius, height):
        self.radius = radius
        self.height = height

    def surface_area(self):
        return 2 * self.pi * self.radius * (self.radius + self.height)
        
    def volume(self):
        return self.pi * self.radius ** 2 * self.height

r, h = map(float, input().split())
c = Cylinder(r, h)
print(f"{c.surface_area():.2f} {c.volume():.2f}")
```

```python
#W8A2 Date
class Date:
    def __init__(self, day, month, year):
        self.day = day
        self.month = month
        self.year = year

    def is_leap(self):
        return (self.year % 400 == 0) or (self.year % 4 == 0 and self.year % 100 != 0)

    def days_in_month(self):
        days = [31, 29 if self.is_leap() else 28, 31, 30, 31, 30,
                31, 31, 30, 31, 30, 31]
        return days[self.month - 1]

    def is_valid(self):
        if self.month < 1 or self.month > 12:
            return False
        if self.day < 1 or self.day > self.days_in_month():
            return False
        return True

    def next_day(self):
        self.day += 1
        if self.day > self.days_in_month():
            self.day = 1
            self.month += 1
            if self.month > 12:
                self.month = 1
                self.year += 1

        return f"{self.day:02d}/{self.month:02d}/{self.year}"

s = input().strip()        
d, m, y = map(int, s.split('/'))

date = Date(d, m, y)
if not date.is_valid():
    print("INVALID")
else:
    print(date.next_day())
```

```python
#W8A3 Point
import math
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def is_origin(self):
        return self.x == 0 and self.y == 0

    def is_y_axis(self):
        return self.x == 0

    def is_x_axis(self):
        return self.y == 0

    def distance_to(self, other):
        return math.sqrt((self.x - other.x) ** 2 + (self.y - other.y) ** 2)

x, y = map(float, input().split())
M = Point(x, y)
O = Point(0, 0)

if M.is_origin():
    print("M la goc toa do")
elif M.is_y_axis():
    print("M nam tren truc tung")
elif M.is_x_axis():
    print("M nam tren truc hoanh")
else:
    print("M khong nam tren truc toa do")
print(f"{M.distance_to(O):.2f}")
```

```python
#W8A4 Calculator
class Calculator:
    def __init__(self, a, b):
        self.a = a
        self.b = b

    def set_numbers(self, a, b):
        self.a = a
        self.b = b

    def cong(self):
        return self.a + self.b

    def tru (self):
        return self.a - self.b

    def nhan (self):
        return self.a * self.b

    def chia (self):
        if self.b == 0:
            return "Khong the chia cho 0"
        return self.a / self.b

    def luy_thua(self):
        return self.a ** self.b

    def chia_du(self):
        if self.b == 0:
            return "Khong the chia du cho 0"
        return self.a % self.b

a, b = map(float, input("Nhap 2 so: ").split())
calc = Calculator(a, b)
while True:
    print("\nMENU")
    print("1. Cong")
    print("2. Tru")
    print("3. Nhan")
    print("4. Chia")
    print("5. Luy thua")
    print("6. Chia du")
    print("7. Nhap 2 so moi")

    choice = input("Chon phep toan (1-7): ")
    if choice == '1':
        print("Ket qua:", calc.cong())
    elif choice == '2':
        print("Ket qua:", calc.tru())
    elif choice == '3':
        print("Ket qua:", calc.nhan())
    elif choice == '4':
        print("Ket qua:", calc.chia())
    elif choice == '5':
        print("Ket qua:", calc.luy_thua())
    elif choice == '6':
        print("Ket qua:", calc.chia_du())
    elif choice == '7':
        a, b = map(float, input("Nhap 2 so moi: ").split())
        calc.set_numbers(a, b)
    else:
        print("Lua chon khong hop le")

    exit_choice = input("Ban co muon thoat? ")
    if exit_choice in ["yes", "YES", "Yes"]:
        break
```

```python
#W8A5 SHopping Cart
class ShoppingCart:
    def __init__(self):
        self.items = [] 

    def add_item(self, name, price):
        self.items.append({"name": name, "price": price})
        
    def remove_item(self, name):
        for item in self.items:
            if item["name"] == name:
                self.items.remove(item)
                return True
        return False

    def is_empty(self):
        return len(self.items) == 0

    def total_price(self):
        return sum(item["price"] for item in self.items)

    def show_items(self):
        if self.is_empty():
            print("Gio hang rong")
        else:
            print("Danh sach san pham:")
            for item in self.items:
                print(f"{item['name']}: {item['price']}")

    def clear_cart(self):
        self.items.clear()

cart = ShoppingCart()
while True:
    print("\nMENU")
    print("2. Them san pham")
    print("3. Xoa san pham")
    print("4. Kiem tra gio rong")
    print("5. Tong tien")
    print("6. Hien thi gio hang")
    print("7. Xoa toan bo gio")

    choice = input("Chon (2-7): ")

    if choice == '2':
        name = input("Ten san pham: ")
        price = float(input("Gia: "))
        cart.add_item(name, price)
        
    elif choice == '3':
        name = input("Nhap ten can xoa: ")
        if cart.remove_item(name):
            print("Da xoa")
        else:
            print("Khong tim thay")

    elif choice == '4':
        print("Gio hang rong" if cart.is_empty() else "Gio hang khong rong")
    elif choice == '5':
        print("Tong tien:", cart.total_price())
    elif choice == '6':
        cart.show_items()
    elif choice == '7':
        cart.clear_cart()
        print("Da xoa toan bo gio")
    else:
        print("Lua chon khong hop le")

    exit_choice = input("Co muon thoat khoi chuong trinh khong? ")
    if exit_choice == "yes":
        break
```

```python
#W8A6 Rectangle
class Rectangle:
    def __init__(self, w, h):
        self.w = w
        self.h = h

    def area(self):
        return self.w * self.h

    def perimeter(self):
        return 2 * (self.w + self.h)

    def scale(self, k):
        self.w *= k
        self.h *= k
        
w, h, k = map(float, input().split())
r = Rectangle(w, h)
r.scale(k)
print(f"{r.area():.2f} {r.perimeter():.2f}")
```

```python
#W8A7 Fraction
import math
class Fraction:
    def __init__(self, num, den):
        if den == 0:
            raise ValueError("Denominator cannot be 0")
        g = math.gcd(num, den)
        num //= g
        den //= g

        if den < 0:   
            num = -num
            den = -den
        self.num = num
        self.den = den

    def add(self, other):
        return Fraction(self.num*other.den + other.num*self.den,
                        self.den*other.den)

    def sub(self, other):
        return Fraction(self.num*other.den - other.num*self.den,
                        self.den*other.den)

    def mul(self, other):
        return Fraction(self.num*other.num, self.den*other.den)

    def div(self, other):
        return Fraction(self.num*other.den, self.den*other.num)

    def __str__(self):
        return f"{self.num}/{self.den}"
        
a, b, op, c, d = input().split()
a, b, c, d = map(int, (a, b, c, d))
f1 = Fraction(a, b)
f2 = Fraction(c, d)

if op == '+':
    res = f1.add(f2)
elif op == '-':
    res = f1.sub(f2)
elif op == '*':
    res = f1.mul(f2)
elif op == '/':
    res = f1.div(f2)
    
print(res)
```

```python
#W8A8 BankAccount
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = float(balance)

    def deposit(self, x):
        self.balance += x

    def withdraw(self, x):
        if x <= self.balance:
            self.balance -= x

    def __str__(self):
        return f"{self.owner}: {self.balance:.2f}"

owner, balance = input().split()
acc = BankAccount(owner, float(balance))
n = int(input())
for _ in range(n):
    cmd, x = input().split()
    x = float(x)
    if cmd == "DEPOSIT":
        acc.deposit(x)
    elif cmd == "WITHDRAW":
        acc.withdraw(x)

print(f"{acc.balance:.2f}")
```

```python
#W8A9 Student
class Student:
    def __init__(self, name):
        self.name = name
        self.scores = []

    def add_score(self, subject, score):
        self.scores.append(score)

    def avg(self):
        return sum(self.scores) / len(self.scores) if self.scores else 0

    def rank(self):
        a = self.avg()
        if a >= 8:
            return "Excellent"
        elif a >= 6.5:
            return "Good"
        elif a >= 5:
            return "Average"
        else:
            return "Poor"

name = input().strip()
n = int(input())
st = Student(name)
for _ in range(n):
    subject, score = input().split()
    st.add_score(subject, float(score))

print(f"{st.name} {st.avg():.2f} {st.rank()}")
```

```python
#W8A10 Book & Library
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = int(year)

class Library:
    def __init__(self):
        self.books = []

    def add(self, book):
        self.books.append(book)

    def count_by_author(self, name):
        return sum(b.author == name for b in self.books)

    def find_by_year(self, y):
        return sum(b.year == int(y) for b in self.books)

lib = Library()
n = int(input())
for _ in range(n):
    line = input()

    if line.startswith("ADD"):
        data = line[4:]  # b·ªè "ADD "
        title, author, year = data.split(';')
        lib.add(Book(title, author, year))

    elif line.startswith("COUNTYEAR"):
        year = line.split()[1]
        print(lib.find_by_year(year))

    elif line.startswith("COUNT"):
        author = line.split(maxsplit=1)[1]
        print(lib.count_by_author(author))
```

```python
#W8A11
import math
def vec(a, b):
    return [b[i] - a[i] for i in range(3)]

def cross(u, v):
    return [
        u[1]*v[2] - u[2]*v[1],
        u[2]*v[0] - u[0]*v[2],
        u[0]*v[1] - u[1]*v[0]
    ]

def dot(u, v):
    return sum(u[i]*v[i] for i in range(3))

def norm(v):
    return math.sqrt(dot(v, v))

A = list(map(float, input().split()))
B = list(map(float, input().split()))
C = list(map(float, input().split()))
D = list(map(float, input().split()))

AB = vec(A, B)
AC = vec(A, C)
n1 = cross(AB, AC)
BC = vec(B, C)
BD = vec(B, D)
n2 = cross(BC, BD)
cos_phi = abs(dot(n1, n2)) / (norm(n1) * norm(n2))
phi = math.degrees(math.acos(cos_phi))
print(f"{phi:.2f}")
```

```python
#W8A12 Complex
class ComplexNumber:
    def __init__(self, realPart, imaginaryPart):
        self.realPart = realPart
        self.imaginaryPart = imaginaryPart

    def multiply(self, other):
        real = self.realPart * other.realPart - self.imaginaryPart * other.imaginaryPart
        imag = self.realPart * other.imaginaryPart + self.imaginaryPart * other.realPart
        return ComplexNumber(real, imag)

r1, i1 = map(float, input().split())
r2, i2 = map(float, input().split())
c1 = ComplexNumber(r1, i1)
c2 = ComplexNumber(r2, i2)
res = c1.multiply(c2)
print(int(res.realPart), int(res.imaginaryPart))
```

```python
#W8A13 Triangle
import math
class Triangle:
    def __init__(self, e1, e2, e3):
        self.edge1 = e1
        self.edge2 = e2
        self.edge3 = e3

    def getArea(self):
        a, b, c = self.edge1, self.edge2, self.edge3

        if a + b <= c or a + c <= b or b + c <= a:
            return -1

        p = (a + b + c) / 2
        return math.sqrt(p * (p-a) * (p-b) * (p-c))
a, b, c = map(int, input().split())
t = Triangle(a, b, c)
area = t.getArea()
if area == -1:
    print("invalid")
else:
    if area.is_integer():
        print(int(area))
    else:
        print(f"{area:.2f}")
```
