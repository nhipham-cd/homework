```python
#W9A1 
def merge(a, b):
    if not a:
        return b
    if not b:
        return a

    if a[0] < b[0]:
        return [a[0]] + merge(a[1:], b)
    else:
        return [b[0]] + merge(a, b[1:])

a = list(map(int, input().split()))
b = list(map(int, input().split()))
print (*merge(a,b))
```

```python
#W9A2
def count_le(matrix, x):
    n, m = len(matrix), len(matrix[0])
    i, j = n - 1, 0
    cnt = 0

    while i >= 0 and j < m:
        if matrix[i][j] <= x:
            cnt += i + 1
            j += 1
        else:
            i -= 1
    return cnt

def kth_largest(matrix, k):
    n, m = len(matrix), len(matrix[0])
    target = n*m - k + 1  
    low, high = matrix[0][0], matrix[-1][-1]

    while low < high:
        mid = (low + high) // 2
        if count_le(matrix, mid) < target:
            low = mid + 1
        else:
            high = mid
    return low

n, m = map(int, input().split())
matrix = [list(map(int, input().split())) for _ in range(n)]
k = int(input())
print(kth_largest(matrix, k))
```

```python
#W9A3
def smallest_swap(n):
    s = list(str(n))
    n = len(s)

    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if s[j] <= s[min_idx]: 
                min_idx = j

        if s[min_idx] < s[i]:
            s[i], s[min_idx] = s[min_idx], s[i]
            break
    return int("".join(s))

num = int(input())
print(smallest_swap(num))
```

```python
#W9A4
def can_convert(a, b):
    n, m = len(a), len(b)
    dp = [[False]*(m+1) for _ in range(n+1)]
    dp[0][0] = True

    for i in range(n):
        for j in range(m+1):
            if not dp[i][j]:
                continue

            if a[i].islower():
                dp[i+1][j] = True

            if j < m and a[i].upper() == b[j]:
                dp[i+1][j+1] = True

    return dp[n][m]

Q = int(input())
for _ in range(Q):
    a = input().strip()
    b = input().strip()
    print("YES" if can_convert(a, b) else "NO")
```

```python
#W9A5
def is_sublist(a, b):
    s = " ".join(map(str, b))
    p = " ".join(map(str, a))
    return p in s
list1 = list(map(int, input().split()))
list2 = list(map(int, input().split()))
print("YES" if is_sublist(list1, list2) else "NO")
```

```python
#W9A7
def canPlaceFlowers(flowerbed, k):
    n = len(flowerbed)
    for i in range(n):
        if flowerbed[i] == 0:
            left = (i == 0 or flowerbed[i-1] == 0)
            right = (i == n-1 or flowerbed[i+1] == 0)

            if left and right:
                flowerbed[i] = 1
                k -= 1

                if k == 0:
                    return True
    return k <= 0

flowerbed = list(map(int, input().split()))
k = int(input())
print(canPlaceFlowers(flowerbed, k))
```

```python
#W9A8
def is_path_crossing(moves):
    x, y = 0, 0
    visited = {(0, 0)}
    for m in moves:
        if m == 'R':
            x += 1
        elif m == 'L':
            x -= 1
        elif m == 'U':
            y += 1
        elif m == 'D':
            y -= 1

        if (x, y) in visited:
            return True
        visited.add((x, y))
    return False
moves = input().strip()
print(is_path_crossing(moves))
```

```python
#W9A9
def Median(nums):
    a, b, c, d, e = nums
    if a > b:
        a, b = b, a
    if c > d:
        c, d = d, c
    if a > c:
        a, c = c, a
        b, d = d, b
    if b > c:
        b, c = c, b
    if c > e:
        c, e = e, c
    if b > c:
        c = b
    return c

nums = list(map(int,input().split()))
print(Median(nums))
```

```python
#W9A10
import bisect
def findRadius(houses, heaters):
    heaters.sort()
    radius = 0
    for h in houses:
        i = bisect.bisect_left(heaters, h)
        left = float('inf')
        right = float('inf')
        if i > 0:
            left = h - heaters[i-1]
        if i < len(heaters):
            right = heaters[i] - h
        radius = max(radius, min(left, right))
    return radius
houses = list(map(int, input().split()))
heaters = list(map(int, input().split()))
print(findRadius(houses, heaters))
```
